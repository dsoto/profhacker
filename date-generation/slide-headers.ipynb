# Generate Slide Headers

This generates a series of markdown slide headers as place holders for future lectures.  I use a string template and generate the dates and headers.

```python
from dates import *
from dateutil import *
from dateutil.rrule import *
import datetime as dt
```

```python
first_day = dt.datetime(2014, 1, 13)
last_day = dt.datetime(2014, 5, 2)
holidays = (dt.datetime(2014, 1, 20),
            dt.datetime(2014, 3, 17),
            dt.datetime(2014, 3, 18),
            dt.datetime(2014, 3, 19),
            dt.datetime(2014, 3, 20),
            dt.datetime(2014, 3, 21),
            dt.datetime(2014, 3, 31)
            )
slidestring = '''## ENSP 202, Lecture {num}
### Professor Daniel Soto, Sonoma State University
### {date:%A, %d %b %Y}
## Topics
## Assigned
## Due
'''
# ENSP 202 meets 10:45-12:00 MW RCH14
output_text(first_day, last_day, (MO, WE), holidays, slidestring)
```
